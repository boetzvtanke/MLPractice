```python
import pandas as pd
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
```

```python
df = pd.read_csv('./data/breast-cancer.csv')
df.head(10)
```

M - рак 

B - доброкачественная опухоль

```python
df.loc[df['diagnosis']=='M', 'diagnosis'] = 0
df.loc[df['diagnosis']=='B', 'diagnosis'] = 1
df['diagnosis'] = pd.to_numeric(df['diagnosis'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='diagnosis'), df['diagnosis'], test_size=0.25)
clsf = CatBoostClassifier(iterations=1000,
                          learning_rate=0.01,
                          depth=10,
                          random_state=42)
clsf.fit(X_train, y_train, verbose=100)
y_true = y_test
y_pred = clsf.predict(X_test)
print("CatBoost:")
print("Метрика F1:", f1_score(y_true, y_pred))
print("Метрика precision:", precision_score(y_true, y_pred))
print("Метрика recall:", recall_score(y_true, y_pred))

clsf = xgb.XGBRFClassifier()
clsf.fit(X_train, y_train)
y_true = y_test
y_pred = clsf.predict(X_test)
print("XGBoost:")
print("Метрика F1:", f1_score(y_true, y_pred))
print("Метрика precision:", precision_score(y_true, y_pred))
print("Метрика recall:", recall_score(y_true, y_pred))

clsf = lgb.LGBMClassifier(num_leaves=31,
                          learning_rate=0.05,
                          n_estimators=100,
                          random_state=42)
clsf.fit(X_train, y_train)
y_true = y_test
y_pred = clsf.predict(X_test)
print("LightGBM:")
print("Метрика F1:", f1_score(y_true, y_pred))
print("Метрика precision:", precision_score(y_true, y_pred))
print("Метрика recall:", recall_score(y_true, y_pred))
```

Лучше работают CatBoost и LightGBM.
